Нахождение гипотенузы прямоугольного треугольника по катетам, нахождение площади треугольника по формуле Герона




```python
import math
import random
katet1 = random.randint(1, 10)
katet2 = random.randint(1, 10)
def gepotinuza(katet1, katet2):
  a = int(katet1)*(int(katet1)) + int(katet2) * (int(katet2))
  c = math.sqrt(a)
  return c
gepotinuza(katet1, katet2)
gep = gepotinuza(katet1, katet2)
print(gep)
p = (katet1 + katet2 + gep) / 2
s = math.sqrt(p*(p-katet1)*(p-katet2)*(p-gep))
print(s)
```
