На дереве есть разные животные. Нужно разнести этих животных по категориям

```python
tree = ['кот', 'голубь', 'воробей', 'ворон', 'гусеница', 'жук', 'белка']
birds = []
animals = []
bugs = []
for i in tree:
  if i == tree[0]:
    animals.append(i)
  elif i == tree[6]:
    animals.append(i)
  elif i == tree[1]:
    birds.append(i)
  elif i == tree[2]:
    birds.append(i)
  elif i == tree[3]:
    birds.append(i)
  elif i == tree[3]:
    birds.append(i)
  elif i == tree[4]:
    bugs.append(i)
  elif i == tree[5]:
    bugs.append(i)
print(birds, animals, bugs)    


```
